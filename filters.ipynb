## OpenCV を使用したフィルタリング例

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('girl.jpg')
kernel = np.ones((5, 5), np.float32) / 25
dst = cv2.filter2D(img, -1, kernel)
plt.imshow(dst)
plt.show()
```

## 平均化フィルター

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('girl.jpg')
dst = cv2.blur(img, (25, 25))
dst1 = dst
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.title('平均化フィルター')
plt.axis('off')
plt.show()
```

## ガウシアンブラー

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('girl.jpg')
dst = cv2.GaussianBlur(img, (25, 25), 0)
dst2 = dst
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.title('ガウシアンブラー')
plt.axis('off')
plt.show()
```

## メディアンフィルター

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('girl.jpg')
dst = cv2.medianBlur(img, 25)
dst3 = dst
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.title('メディアンフィルター')
plt.axis('off')
plt.show()
```

## 微分フィルター

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('girl.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
dstx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
dsty = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
dst = cv2.magnitude(dstx, dsty)
dst4 = dst
plt.imshow(dst, cmap='gray')
plt.title('微分フィルター')
plt.axis('off')
plt.show()
```

## ラプシアンフィルター

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('girl.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
dst = cv2.Laplacian(img, cv2.CV_64F)
dst5 = dst
display(dst)
plt.imshow(np.abs(dst), cmap='gray')
plt.title('ラプシアンフィルター')
plt.axis('off')
plt.show()
```

## アンシャープマスキング

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('girl.jpg')
dst = cv2.addWeighted(img, 1.5, dst2, -0.5, 0)
dst6 = dst
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.title('アンシャープマスキング')
plt.axis('off')
plt.show()
```

## オリジナル画像とアンシャープマスキング画像の比較

`jupyter環境で、画像※フィルタリング処理前後の結果を比較して、処理後の画像が良くなった理由を説明してください。20点`

アンシャープマスキングは画像のエッジや輪郭を強調する効果がある。処理後の画像では髪の毛の細かい部分や帽子の輪郭がよりシャープになっている。特に目や口、肌の質感などの顔の特徴がより鮮明に表現され、人物の表情がはっきりと認識できるようになっている。

処理により、元の画像で少しぼやけていた細部が明確になり、髪の毛一本一本や帽子の編み目などの細かい模様がよりはっきりと見えるようになっている。この細部の回復により、画像全体のディテールが豊かになり、より高品質な印象を与える。

さらに、アンシャープマスキングは局所的なコントラストを高める効果があるため、明暗の差が強調され画像全体の立体感が増している。これにより、処理後の画像は全体的にクリアで焦点が合っているように見え、視覚的な鮮明さと奥行き感が向上している。
